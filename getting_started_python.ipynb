### Dictionaries


```python
meal = {"breakfast":"Idly", "lunch":"Burger", "dinner": "Roti"}
meal.keys()
```

```python
meal.values()
```

```python
for meal_name, meal_item in meal.items():
  print(f"This meal {meal_name} is when I like to Consume {meal_item}")
```
